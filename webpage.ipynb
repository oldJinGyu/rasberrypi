```python
from flask import Flask, render_template, redirect, url_for
import RPi.GPIO as GPIO
import time

GPIO.setmode(GPIO.BCM)

sensor = 21
sound = 18
GPIO.setup(sound, GPIO.OUT)
state = 0

app = Flask(__name__)

@app.route('/')
def cctv():
  return render_template('web.html',state = state)

@app.route('/<int:s>')
def control(s):
  state = s
  for i in range(20):
    GPIO.output(sound, state)
    time.sleep(0.25)
    GPIO.output(sound, 0)
    time.sleep(0.25)

  return redirect(url_for('cctv'))

if __name__ == '__main__':
  app.run(host='0.0.0.0')
```
